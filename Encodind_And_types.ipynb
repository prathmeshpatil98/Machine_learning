## Nominal | One Hot Encoding

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder


df = pd.DataFrame({
    'color':['red','blue', 'green', 'green', 'red', 'blue']
})


df
```

```python
encoder = OneHotEncoder()
```

```python
encoder.fit_transform(df[['color']]).toarray()
```

```python
encoded_df = pd.DataFrame(encoder.fit_transform(df[['color']]).toarray(), columns=encoder.get_feature_names_out())


encoded_df
```

```python
pd.concat([df, encoded_df], axis=1)
```

```python

pd.concat([df, encoded_df])
```

## Label And Ordinal Encoding

```python

import pandas as pd
from sklearn.preprocessing import LabelEncoder


# create a sample dataframe with a categorical variable
df = pd.DataFrame({
    'color':['red','blue', 'green', 'green', 'red', 'blue']
})


df
```

```python

# create an instance of labelencoder
encoder = LabelEncoder()


encoder.fit_transform(df['color'])
```

```python
# Ordinal Encoding
from sklearn.preprocessing import OrdinalEncoder

# create a sample dataframe with a categorical variable
df = pd.DataFrame({
    'size':['small','medium', 'large', 'medium', 'small', 'large']
})
```

```python
list(df['size'].unique())
```

```python
unique_sizes = df['size'].unique()


# create an instance of the ordinalEncoder class and fit_transform
encoder = OrdinalEncoder(categories=[unique_sizes])


encoder.fit_transform(df[['size']])
```

```python
# create an instance of the ordinalEncoder class and fit_transform
encoder = OrdinalEncoder(categories=[unique_sizes])


encoder.fit_transform(df[['size']])
```

## Target Guided Ordinal Encoding

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'city' : ['new york','London','paris','Tokyo','new york','paris'],
    'price': [200,150,300,250,180,320]})
```

```python
df
```

```python
df.groupby('city')['price']
```

```python
df.groupby('city')['price'].mean()
```

```python
mean_price = df.groupby('city')['price'].mean().to_dict()
```

```python
mean_price
```

```python
df['city_Embeded'] = df['city'].map(mean_price)
```
