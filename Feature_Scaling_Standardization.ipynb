## Standardization

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df['total_bill']
```

```python
total_bill = list(df['total_bill'])
```

```python
import numpy as np
mean = np.mean(total_bill)
std = np.std(total_bill)
mean , std
```

```python
normalized_Data = []
for i in total_bill:
  z_score = (i-mean)/std
  normalized_Data.append(z_score)
print(normalized_Data)
```

```python
sns.histplot(total_bill)
```

```python
sns.displot(normalized_Data)
```

# Using SKlearn

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler
```

```python
scaler.fit(df[['total_bill' , 'tip']])
```

```python
scaler.transform(df[['total_bill' , 'tip']])
```

```python
import pandas as pd
df1 = pd.DataFrame(scaler.transform(df[['total_bill' , 'tip']]),columns=['total_bills','tip1'])
```

```python
df1
```

```python
sns.histplot(df1['total_bills'])
```

```python
sns.histplot(df1['tip1'])
```
